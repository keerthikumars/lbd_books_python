```python
import pandas as pd
import datetime as dt
from pandas_datareader import data
```

```python
companies = ["MSFT","GOOG","AAPL","ERIC","AMZN"]
start   = "2013-01-01"
end     = "2018-03-31"
d = data.DataReader(name=companies, data_source="iex", start=start, end=end)
t = pd.Panel.from_dict(d,orient="minor")
```
